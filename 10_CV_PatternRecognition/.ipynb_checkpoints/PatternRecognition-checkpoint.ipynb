#### Introduccion Procesamiento de Imagenes

Vamos a querer procesar mejorando o generando una mejor imagen o nueva imagen. 
Desde el punto de vista de la computadora, la imagen es un arrray de pixeles. Computer Vision trata de querer tomar decisiones relevantes para nuestra aplicacion a partir del array de pixeles, el cual es una matriz de 2 dimesiones donde cada celda puede tener una dimension mas dependiendo del formato que se le de.

#### Cargando y Mostrando imagenes

Se importa el paquete ``mahotas`` el cual esta basado en NumPy.

```python
import mahotas as mh
from matplotlib import pyplot as plt

image = mh.imread('a.jpg') # image es un np.array
plt.imshow(image)
plt.show()

image = image - image.mean() # Esto ayuda a normlizar la images
```

```python
plt.imshow(image)
plt.show()
```

Si ``a.jpg`` es una imagen a color de altura ``h`` y ancho ``w``, entonces la imagen va a ser un array de la forma (h, w, 3). La tercera dimension vendria a ser los colores **RGB** (red, green, blue). Esto ultimo es la convencion usada en NumPy. El tipo de dato va a ser np.uint8 (unsigned integer 8 bits).
